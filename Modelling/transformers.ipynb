```python
import os
paths = ['/content/drive/MyDrive/SPIRE/Modelling/espnet/egs2/TEMPLATE/asr1/dump/test/test.txt', '/content/drive/MyDrive/SPIRE/Modelling/espnet/egs2/TEMPLATE/asr1/dump/dev/dev.txt', '/content/drive/MyDrive/SPIRE/Modelling/espnet/egs2/TEMPLATE/asr1/dump/train/train.txt']

for filename in paths:
    with open(filename, 'r') as f:
        lines = f.read().split('\n')[:-1]
    lines = [f'id{str(idx)} {l}\n' for idx, l in enumerate(lines)]
    with open(filename, 'w') as f:
        for l in lines:
            f.write(l)
```
