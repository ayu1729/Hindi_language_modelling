```python
#loading the corpus
import numpy as np
corpus=np.load('/content/drive/MyDrive/SPIRE/Dataset/Data_csv/Final/Corpus_no_eng_no_punc.npy',allow_pickle=True)#address of the npy file where sentences are stored
```

```python
corpus=list(corpus)
```

```python
#splitting into train and test set
from sklearn.model_selection import train_test_split
train, test = train_test_split(corpus, test_size = 0.2)
```

Unigram without smoothing

```python
#padding
from nltk.lm.preprocessing import pad_both_ends
test=[list(pad_both_ends(t, n=1)) for t in test]
train=[list(pad_both_ends(t, n=1)) for t in train]
```

```python
#create_vocabulary from the train dataset with cutoff of 2
words=[]
for i in train:
  for j in i:
    words.append(j)
from nltk.lm import Vocabulary
vocab = Vocabulary(words, unk_cutoff=2)
```

```python
"UNK" in vocab
```

```python
#convert data to n-gram sequence
from nltk.util import ngrams
new_test=[list(ngrams(t,1)) for t in test]
new_train=[list(ngrams(t,1)) for t in train]
```

```python
train_len=len(train)
test_len=len(test)
```

```python
train_len #no of sentences in training set
```

```python
test_len #no of sentences in training set
```

```python
from nltk.lm import MLE#using maximum likelihood estimation
lm = MLE(1)
```

```python
lm.fit(new_train,vocab)#training the model
```

```python
"UNK" in lm.vocab
```

```python
lm.score("UNK")#checking score on unknown token making sure it's not zero
```

```python
perplexity_list=[]
```

```python
for t in new_test:
  perplexity_list.append(lm.perplexity(t))
```

```python
sum=0
for p in perplexity_list:
  sum+=p
```

```python
avg=sum/test_len
```

```python
import numpy as np
import math
overall_perplexity=math.pow(2,np.sum(np.log2(np.array(perplexity_list)))/test_len)
```

```python
from nltk.lm.preprocessing import pad_both_ends
test=[list(pad_both_ends(t, n=4)) for t in test]
train=[list(pad_both_ends(t, n=4)) for t in train]
```

```python
words=[]
for i in train:
  for j in i:
    words.append(j)
from nltk.lm import Vocabulary
vocab = Vocabulary(words, unk_cutoff=2)
```

```python
"UNK" in vocab
```

```python
from nltk.util import ngrams
new_test=[list(ngrams(t,4)) for t in test]
new_train=[list(ngrams(t,4)) for t in train]
```

```python
"UNK" in vocab
```

```python
from nltk.lm import KneserNeyInterpolated#using maximum likelihood estimation
lm = KneserNeyInterpolated(4,0.1)
```

```python
lm.fit(new_train,vocab)
```
